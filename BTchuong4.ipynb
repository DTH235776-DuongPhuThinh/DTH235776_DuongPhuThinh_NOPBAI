```python
#Câu 1: Viết hàm để Tính diện tích tam giác
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam giác không hợp lệ")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Diện tích =",dt)
```

```python
#Câu 2: Viết Hàm để chơi Game Đoán Số
from random import randrange

while True:
    somay = randrange(1, 101)
    solandoan = 0
    win = False
    while solandoan < 7:
        solandoan += 1
        songuoi = int(input("Máy ra số [1..100], mời bạn đoán: "))
        print("Bạn đoán lần thứ ", solandoan)
        if somay == songuoi:
            print("Chúc mừng bạn đã đoán đúng! Số máy là: ", somay)
            win = True
            break
        elif somay > songuoi:
            print("Bạn đoán sai, số máy lớn hơn số bạn đoán")
        else:
            print("Bạn đoán sai, số máy nhỏ hơn số bạn đoán")
    if not win:
        print("Bạn đã thua! Số máy là: ", somay)
    hoi = input("Tiếp tục? (k để thoát): ")
    if hoi.lower() == 'k':
        break
print("Cám ơn bạn đã chơi game!")
```

```python
#Câu 3: Viết Hàm tính BMI
def BMI(height, weight):
    bmi = weight / (height ** 2)
    return weight / (height ** 2)

def PhanLoai(bmi):
    if bmi < 18.5:
        return "Gầy"
    elif bmi <= 24.9:
        return "Bình thường"
    elif bmi <= 29.9:
        return "Hơi béo"
    elif bmi <= 34.9:
        return "Béo phì cấp độ 1"
    elif bmi <= 39.9:
        return "Béo phì cấp độ 2"
    else:
        return "Béo phì cấp độ 3"
    
def NguyCoBenh(bmi):
    if bmi < 18.5:
        return "Thấp"
    elif bmi <= 24.9:
        return "Trung bình"
    elif bmi <= 29.9:
        return "Cao"
    elif bmi <= 34.9:
        return "Rất cao"
    elif bmi <= 39.9:
        return "Nguy hiểm"
    else:
        return "Cực kỳ nguy hiểm"

print("Nhập vào chiều cao (m): ")
height = float(input())
print("Nhập vào cân nặng (kg): ")
weight = float(input())

bmi = BMI(height, weight)
print("BMI của bạn: ", bmi)
print("Phân loại: ", PhanLoai(bmi))
print("Nguy cơ mắc bệnh: ", NguyCoBenh(bmi))
```

```python
#Câu 4: Viết Hàm tính ROI
def ROI(dt,cp):
 return (dt-cp)/cp
def GoiYDauTu(roi):
 if roi>=0.75:
    return "Nên đầu tư"
 else:
    return "Không nên đầu tư"
print("Chương trình tính ROI")
dt=int(input("Nhập Doanh Thu:"))
cp=int(input("Nhập chi phí:"))
roi=ROI(dt,cp)
print("Tỉ Lệ ROI=",roi)
print("==>",GoiYDauTu(roi))
```

```python
#Câu 5: Viết hàm đệ qui Fibonacci
def fibonacci(n):
 if n<=2 :
    return 1
 return fibonacci(n-1)+fibonacci(n-2)
def listfibo(n):
 for i in range(1,n+1):
    print(fibonacci(i),end='\t')
print(fibonacci(9))
listfibo(9)
```

```python
#Câu 7: Tính và xuất độ dài đoạn AB
from math import sqrt

xA = float(input("Nhập hoành độ điểm A: "))
yA = float(input("Nhập tung độ điểm A: "))
xB = float(input("Nhập hoành độ điểm B: "))
yB = float(input("Nhập tung độ điểm B: "))

AB = sqrt((xB - xA)**2 + (yB - yA)**2)
print("Độ dài đoạn AB là:", AB)
```

```python
#Câu 8: Viết chương trình tính log(a)^x
from math import log
a = float(input("Nhập cơ số a (a>0 và a≠1): "))
x = float(input("Nhập số mũ x: "))

if a <= 0 or a == 1 or x < 0:
    print("Giá trị không hợp lệ.")
else:
    loga_x = log(x) / log(a)
    print(f"log cơ số {a} của {x} là: {loga_x}")
```

```python
#Câu 9: Viết chương trình tính căn bậc 2 lồng nhau
from math import sqrt

n = int(input("Nhập số lần căn bậc 2 lồng nhau (n): "))
s = 0
for i in range(n):
    s = sqrt(2 + s)
print(f"Kết quả của căn bậc 2 lồng nhau sau {n} lần là: {s}")
```
