```python
#Câu 1: Xử lý List
from random import randrange

print("Chương trình xử lý List")
n = int(input("Nhập số phần tử của list: "))
lst = [randrange(-100, 100) for _ in range(n)]

print("List ngẫu nhiên:", lst)
value = int(input("Mời bạn thêm số mới: "))
lst.append(value)
print("List sau khi thêm:", lst)

k = int(input("Nhập số k cần đếm: "))
dem = lst.count(k)
print(f"Số {k} xuất hiện {dem} lần trong list.")

def CheckPrime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

tongnt = sum(x for x in lst if CheckPrime(x))
demnt = sum(1 for x in lst if CheckPrime(x))
print("Có", demnt, "số nguyên tố trong list với tổng là:", tongnt)

lst.sort()
print("List sau khi sắp xếp:", lst)

del lst
print("List đã được xóa.")
```

```python
#Câu 2: Xử lý List nhập ngẫu nhiên
from random import randrange

lst = []
n = int(input("Nhập số phần tử của list: "))
for i in range(n):
    lst.append(randrange(0, 100))

print("List ngẫu nhiên:", lst)
x = int(input("Mời bạn chèn thêm số mới: "))
lst.append(x)
print("List sau khi thêm:", lst)

k = int(input("Nhập số để xóa: "))
while k in lst:
    lst.remove(k)
print("List sau khi xóa:", lst)

def CheckDoiXung(lst):
    for i in range(len(lst)):
        if lst[i] != lst[len(lst) - i - 1]:
            return False
    return True
if CheckDoiXung(lst):
    print("List là đối xứng.")
else:
    print("List không là đối xứng.")
```

```python
#Câu 3: Xử lý List Đa chiều
from random import randrange

def TaoMaTran(m, n):
    return [[randrange(100) for _ in range(n)] for _ in range(m)]

def XuatMaTran(D):
    for row in D:
        print('\t'.join(map(str, row)))

def LayDong(D, r):
    return D[r]

def LayCot(D, c):
    return [D[i][c] for i in range(len(D))]

def TimMax(D):
    return max(max(row) for row in D)

m = int(input("Nhập số dòng của ma trận: "))
n = int(input("Nhập số cột của ma trận: "))
D = TaoMaTran(m, n)
XuatMaTran(D)

r = int(input("Nhập dòng cần xuất: "))
print("Dòng", r, ":", LayDong(D, r))

c = int(input("Nhập cột cần xuất: "))
print("Cột", c, ":", LayCot(D, c))

print("Số lớn nhất trong ma trận là:", TimMax(D))
```

```python
#Câu 10: Xử lý Ma Trận
m = int(input("Nhập số dòng của ma trận: "))
n = int(input("Nhập số cột của ma trận: "))

print("Nhập ma trận A:")
A = [[float(input(f"A[{i}][{j}] = ")) for j in range(n)] for i in range(m)]
print("Nhập ma trận B:")
B = [[float(input(f"B[{i}][{j}] = ")) for j in range(n)] for i in range(m)]

C = [[A[i][j] + B[i][j] for j in range(n)] for i in range(m)]
print("Ma trận A + B là: ", C)

AT = [[A[i][j] for i in range(m)] for j in range(n)]
BT = [[B[i][j] for i in range(m)] for j in range(n)]
print("Hoán vị ma trận A là: ", AT)
print("Hoán vị ma trận B là: ", BT)
```
