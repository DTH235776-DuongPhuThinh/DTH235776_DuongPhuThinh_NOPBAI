```python
#Câu 1: Kiểm tra chuỗi đối xứng
def CheckDoiXung(s):
 flag=True
 for i in range(len(s)):
    if s[i]!=s[len(s)-i-1]:
        flag=False
        break
 return flag
def main():
 print("Nhập 1 chuỗi:")
 s=input()
 if(CheckDoiXung(s)):
    print("Chuỗi bạn nhập đối xứng")
 else:
    print("Chuỗi bạn nhập không đối xứng")
while True:
 main()
 print("Tiếp không Thím?(c/k):")
 s=input()
 if s=="k":
    break
print("CÁM ƠN THÍM")
```

```python
#Câu 2: Viết chương trình tối ưu chuỗi
def ToiUuChuoi(s):
 s2=s
 s2=s2.strip()
 arr=s2.split(' ')
 s2=""
 for x in arr:
    word=x
 if len(word.strip())!=0:
    s2=s2+word+" "
 return s2.strip()
s=" Trần Duy Thanh "
print(s,"=>",len(s))
s=ToiUuChuoi(s)
print(s,"=>",len(s))
```

```python
#Câu 6: Trích lọc số âm trong chuỗi
import re

def NegativeNumberInStrings(s):
    return re.findall(r'-\d+', s)
s = "abc-5xyz-12k9l--p"
print("Các số âm trong chuỗi là:", NegativeNumberInStrings(s))
```

```python
#Câu 7: Tối ưu chuỗi danh từ
def ToiUuDanhTu(s):
    s = " ".join(s.split())
    words = s.lower().split()
    return " ".join([w.capitalize() for w in words])

s = "   trẦn    duy    thANH   "
print("Chuỗi tối ưu:", ToiUuDanhTu(s))
```
